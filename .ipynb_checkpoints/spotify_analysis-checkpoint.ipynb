```python
df = pd.read_excel('Spotify_data.xlsx')

df = clean_data(df)
```

```python
from scipy.stats import chi2_contingency

df_transformed = df.apply(lambda x: x.astype('category').cat.codes if x.dtype == 'object' else x)

print("Transformed DataFrame:")
print(df_transformed)

# Chi-Square test for independence
target = 'premium_sub_willingness'
print(f"\nTarget Variable Distribution:\n{df[target].value_counts()}\n")

for column in df_transformed.columns:
    if column != target:
        contingency_table = pd.crosstab(df_transformed[column], df_transformed[target])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        print(f"Chi-Square Test for {column} and {target}:")
        print(f"Null Hypothesis: There is no association between {column} and {target}.")
        print(f"Alternative Hypothesis: There is an association between {column} and {target}.")
        print(f"Chi2 = {chi2}, p-value = {p}\n")
        
        if p < 0.05:
            print("Conclusion: Reject the null hypothesis. There is a significant association between", column, "and", target + ".\n")
        else:
            print("Conclusion: Fail to reject the null hypothesis. There is no significant association between", column, "and", target + ".\n")
```

```python
target = 'premium_sub_willingness'
print(f"\nTarget Variable Distribution:\n{df[target].value_counts()}\n")
p_values = []
for column in df_transformed.columns:
    if column != target:
        contingency_table = pd.crosstab(df_transformed[column], df_transformed[target])
        chi2, p, _, _ = chi2_contingency(contingency_table)
        p_values.append(p)

# Create a dataframe for p-values
p_values_df = pd.DataFrame([p_values], columns=df_transformed.columns[:-1], index=['p-value'])

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(p_values_df, annot=True, cmap="coolwarm", fmt=".3f", linewidths=.5)
plt.title('Chi-Square p-values for Feature vs Premium Subscription Willingness')
plt.xlabel('Features')
plt.ylabel('p-value')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
