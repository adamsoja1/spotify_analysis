```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

df = pd.read_excel('Spotify_data.xlsx')
```

```python
df_encoded = pd.get_dummies(df.drop('premium_sub_willingness', axis=1))
df_encoded['premium_sub_willingness'] = df['premium_sub_willingness']
```

```python
df = df.drop('preffered_premium_plan', axis=1)
#df = df.drop('spotify_subscription_plan', axis=1)
df = df.dropna()

Y = df_encoded['premium_sub_willingness']
df = df_encoded.drop(columns=['premium_sub_willingness'])
```

```python
for col in df.columns:
    df[col] = df[col].astype(float)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(df, Y, test_size=0.2, random_state=0, stratify=Y)
```

```python
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X_train, y_train)
y_pred_rf = rf_clf.predict(X_test)
rf_accuracy = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Classifier Accuracy: {rf_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_rf))
```

```python
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
log_reg_accuracy = accuracy_score(y_test, y_pred_log_reg)
print(f"Logistic Regression Accuracy: {log_reg_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_log_reg))
```

```python
svc = SVC()
svc.fit(X_train, y_train)
y_pred_svc = svc.predict(X_test)
svc_accuracy = accuracy_score(y_test, y_pred_svc)
print(f"Support Vector Classifier Accuracy: {svc_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_svc))
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_pred_knn)
print(f"K-Nearest Neighbors Classifier Accuracy: {knn_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_knn))
```

```python
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train, y_train)
y_pred_dt = dt_clf.predict(X_test)
dt_accuracy = accuracy_score(y_test, y_pred_dt)
print(f"Decision Tree Classifier Accuracy: {dt_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_dt))
```

```python
gb_clf = GradientBoostingClassifier(random_state=42)
gb_clf.fit(X_train, y_train)
y_pred_gb = gb_clf.predict(X_test)
gb_accuracy = accuracy_score(y_test, y_pred_gb)
print(f"Gradient Boosting Classifier Accuracy: {gb_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_gb))

# AdaBoost Classifier
ada_clf = AdaBoostClassifier(random_state=42)
ada_clf.fit(X_train, y_train)
y_pred_ada = ada_clf.predict(X_test)
ada_accuracy = accuracy_score(y_test, y_pred_ada)
print(f"AdaBoost Classifier Accuracy: {ada_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_ada))

# Naive Bayes Classifier
nb_clf = GaussianNB()
nb_clf.fit(X_train, y_train)
y_pred_nb = nb_clf.predict(X_test)
nb_accuracy = accuracy_score(y_test, y_pred_nb)
print(f"Naive Bayes Classifier Accuracy: {nb_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_nb))
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 200, 300, 400],
    'learning_rate': [0.01, 0.1, 0.2, 0.3, 0.6],
    'max_depth': [3, 4, 5, 6, 7, 8]
}
gb_classifier = GradientBoostingClassifier()
grid_search = GridSearchCV(estimator=gb_classifier, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
print("\nBest Hyperparameters:", grid_search.best_params_)
y_pred = best_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy}")
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
best_params = None
best_score = 0
for n_estimators in param_grid['n_estimators']:
    for learning_rate in param_grid['learning_rate']:
        for max_depth in param_grid['max_depth']:
            gb_classifier = GradientBoostingClassifier()
            # Set the parameters
            gb_classifier.set_params(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth)
            # Train the model
            gb_classifier.fit(X_train, y_train)
            # Evaluate the model on the validation set
            score = gb_classifier.score(X_test, y_test)
            print(f"Params: n_estimators={n_estimators}, learning_rate={learning_rate}, max_depth={max_depth}, Score: {score}")
            if score > best_score:
                best_score = score
                best_params = {
                    'n_estimators': n_estimators,
                    'learning_rate': learning_rate,
                    'max_depth': max_depth
                }

print("\nBest Hyperparameters:", best_params)
```

```python
gb_clf = GradientBoostingClassifier(random_state=42, learning_rate=0.01, max_depth=3, n_estimators=200)
gb_clf.fit(X_train, y_train)
y_pred_gb = gb_clf.predict(X_test)
gb_accuracy = accuracy_score(y_test, y_pred_gb)
print(f"Gradient Boosting Classifier Accuracy: {gb_accuracy:.2f}")
print(confusion_matrix(y_test, y_pred_gb))
```
